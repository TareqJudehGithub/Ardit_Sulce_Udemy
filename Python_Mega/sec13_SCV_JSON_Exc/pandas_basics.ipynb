```python
import pandas
```

```python
# pandas converts data files into a table format.
data = pandas.read_csv('data.txt')
data
```

```python
# As you noticed above, the data.txt table has header, so lets fix that, shall we?
data.columns = ["ID", "Address", "City", "Zip", "Country", "Customer", "Employees"]
data
# data.columns returns the table columns
```

```python
# setting ID as the table index
data.set_index("ID", inplace=True)

# data.index returns the table indexes
```

```python
data
```

```python
# Data filtering in pandas:
# 1. using labels
# get me the first three customers
customers_label = data.loc[0:3 , "Customer"]
customers_label
```

```python
# 2. using indexing(slicing)
# get me the first three addresses
address_indexing = data.iloc[0:3, 0]
address_indexing

```

```python
# delete rows and columns
# to delete a column, provide the desired column name in the first argument, and 1 in the second.
# data.drop("column", 1)

# to delete a row, provide the desired column name in the first argument, and 0 in the second.
# data.drop("row", 0)
```

```python
# delete the Employees column
data.drop("Employees", 1)
data
```

```python
# delete the last record from the table
data.drop(data.index[-1], 0)
```
